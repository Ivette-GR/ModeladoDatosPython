```python
import pandas as pd
import matplotlib.pyplot as plt 
import folium
from folium.plugins import HeatMap

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.head()
```

*Ingreso total por cada tienda
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.


```python
SumaTienda=sum(tienda.Precio);
SumaTienda2=sum(tienda2.Precio);
SumaTienda3=sum(tienda3.Precio);
SumaTienda4=sum(tienda4.Precio);
print (SumaTienda);
print (SumaTienda2);
print (SumaTienda3);
print(SumaTienda4);
```

Ventas por categoría
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
tienda['tienda']="Tienda1"; #Agrego una columna identificando a la tienda que pertenece
tienda.head()
```

```python
tienda2['tienda']="Tienda2"; #Agrego una columna identificando a la tienda que pertenece
tienda2.head()
```

```python
tienda3['tienda']="Tienda3"; #Agrego una columna identificando a la tienda que pertenece
tienda3.head()
```

```python
tienda4['tienda']="Tienda4"; #Agrego una columna identificando a la tienda que pertenece
tienda4.head()
```

```python
BaseCompleta=pd.concat([tienda,tienda2,tienda3,tienda4]);
BaseCompleta.info()
```

```python
conteo = BaseCompleta.groupby(['tienda', 'Categoría del Producto']).size().reset_index(name='ventas').sort_values(by= 'ventas', ascending=False);
conteo.head(10)
```

Valoración media por tienda
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
promedios= BaseCompleta.groupby(['tienda'])['Calificación'].mean().reset_index()
print(promedios);
```

Valor del envío promedio por tienda
En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
promedioEnvio= BaseCompleta.groupby(['tienda'])['Costo de envío'].mean().reset_index()
print(promedioEnvio);
```

Graficos

```python
pivote=conteo.pivot(index='Categoría del Producto', columns='tienda', values='ventas')
pivote.head()
```

```python
pivote.plot(kind='bar', title='Ventas por producto y tienda')
plt.ylabel('ventas')
plt.xlabel('Producto')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
conteo2 = BaseCompleta.groupby('tienda')['Precio'].sum().reset_index().sort_values(by='Precio', ascending=False)
conteo2.head(10)

```

```python
conteo2.plot(kind='bar', title='Ventas por producto')
plt.ylabel('Precio')
plt.xlabel('tienda')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Crear el mapa centrado en la ubicación media
mapa = folium.Map(location=[BaseCompleta['lat'].mean(), BaseCompleta['lon'].mean()], zoom_start=13)

# Crear lista de puntos para el mapa de calor
puntos = BaseCompleta[['lat', 'lon', 'Calificación']].values.tolist()

# Agregar capa de calor
HeatMap(puntos).add_to(mapa)

# Mostrar mapa (en Jupyter)
mapa
```
